```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats as stats
import seaborn as sns
```

```python
# generate data
size = 200
x = stats.norm.rvs(size=size)
ejects = np.array([5, -4, 3.3, 2.99, -3])
for i, eject in enumerate(ejects):
    x[i] = eject
x_sorted = np.array(sorted(x))
print(x)
```

```python
# 3sig rull
def sigm3_rull(x: np.array):

    xmean = np.mean(x)
    xvar = np.var(x)
    is_in_3sigm = lambda xi: np.abs(xi - xmean) < 3*xvar

    clear_res = [val for i, val in enumerate(x) if is_in_3sigm(val)]
    clear_res_x = [i for i, val in enumerate(x) if is_in_3sigm(val)]

    return clear_res_x, clear_res

clear_sample = sigm3_rull(x)

plt.plot(x, '*')
plt.plot(*clear_sample)
plt.grid()

ej_sample = [val for val in x if val not in clear_sample[1]]
print(ej_sample)
```
