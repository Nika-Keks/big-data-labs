```python
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

```python
#2
place_metrics = {
    "mean": np.mean,
    "median": np.median,
    "perceptile sub": lambda x: (np.percentile(x, 75) - np.percentile(x, 25) / 2.),
    "extrim sub": lambda x: (np.max(x) - np.min(x) / 2.)
}
cauchy_sample = stats.cauchy.rvs(size=100)
for name, metric in place_metrics.items():
    print(f"{name} = {metric(cauchy_sample)}")    
```

```python
#3
uniform_sample = stats.uniform.rvs(size=100)
plt.hist(uniform_sample, density=True)
x = np.arange(-1., 2., 0.01)
plt.plot(x, stats.uniform.pdf(x=x, loc=0, scale=1))
plt.grid()
```

```python
#4
laplace_sample = sorted(stats.laplace.rvs(size=100))
x = np.linspace(1 / len(laplace_sample), 1, len(laplace_sample))
plt.step(laplace_sample, x)
plt.plot(laplace_sample, stats.laplace.cdf(laplace_sample))
plt.grid()
```
